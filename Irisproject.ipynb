```python
X = iris.iloc[:,0:3]
X
```

```python
y = iris.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train , y_train)
```

```python
y_predict = model.predict(X_test)
print(y_predict)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_predict = model.predict(X_test)
accuracy_score (y_test , y_predict)
```

```python
import numpy as np
from mlxtend.plotting import plot_decision_regions
from sklearn.base import clone


X_v = X_train.values if hasattr(X_train, 'values') else X_train
X_plot = X_v[:, :2] 


if hasattr(y_train, 'values'):
    
    y_plot = y_train.astype('category').cat.codes.values
else:
    
    y_plot = y_train.astype(int)

model_2d = clone(model)
model_2d.fit(X_plot, y_plot)

plt.figure(figsize=(8, 5))
plot_decision_regions(X_plot, y_plot, clf=model_2d, legend=2)

plt.title(f"Decision Regions (Acc: 0.93)")
plt.show()
```
